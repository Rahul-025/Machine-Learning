```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Function to minimise :

### $$ f(x) = 4x^2 + 5x  $$
### $$ df(x) = 8x + 5 $$

```python
def f(x):
    return 4*x**2 + 5*x

def df(x):
    return 8*x + 5
```

```python
x = np.linspace(start=-3, stop=3 ,num=1000)
```

### Plotting the function and its derivative:

```python
# Function Plot :

plt.figure(figsize=(15, 5))

plt.subplot(1, 2 ,1)
plt.title('Function Plot', fontsize=12)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('y-axis', fontsize=12)
plt.xlim(-3, 3)
plt.ylim(-4, 60)
plt.plot(x, f(x), color='violet', linewidth=5)

# Derivative :

plt.subplot(1, 2 ,2)
plt.title('Derivative PLot', fontsize=12)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('y-axis', fontsize=12)
plt.xlim(-3, 3)
plt.plot(x, df(x), color='pink', linewidth=5)
```

### Gradient Descent function :

```python
prev_x = 0
def gradient_descent(x_current, max_iter=200, learning_rate=0.1):
    x_list = [x_current]
    slope_list = [df(x_current)]
    for i in range(max_iter):
        prev_x = x_current
        slope = df(prev_x) 
        x_current = prev_x - learning_rate * slope
        
        x_list.append(x_current)
        slope_list.append(df(x_current))
                          
        if(abs(x_current - prev_x) < 0.0001):
            break
    return x_current, x_list, slope_list
```

### Plotting Gradient Descent

```python
minima, list_x, list_slope = gradient_descent(x_current=2, max_iter=200, learning_rate=0.05)
print('The minima occurs at : ', minima)
print('The Cost at minima is : ', f(minima))
print('The Slope at minima is : ', df(minima))

# Plotting the x values on our Function Plot :

plt.figure(figsize=(15, 5))

plt.subplot(1, 2 ,1)
plt.title('Function Plot', fontsize=12)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('y-axis', fontsize=12)
plt.xlim(-3, 3)
plt.ylim(-4, 55)
plt.plot(x, f(x), color='violet', linewidth=5, alpha=0.7)
plt.scatter(list_x, f(np.array(list_x)), s=100, color='red', alpha=0.6)


# Plotting the x values on our Derivative Plot :

plt.subplot(1, 2 ,2)
plt.title('Derivative PLot', fontsize=12)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('y-axis', fontsize=12)
plt.xlim(-3, 3)
plt.plot(x, df(x), color='pink', linewidth=5, alpha=0.7)
plt.scatter(list_x, list_slope, s=100, color='red', alpha=0.6)
```

### Effect of learning rate on function :

```python
n = 100
low_learning_rate = gradient_descent(x_current=2.2, max_iter=n, learning_rate=0.001)
mid_learning_rate = gradient_descent(x_current=2.2, max_iter=n, learning_rate=0.002)
high_learning_rate = gradient_descent(x_current=2.2, max_iter=n, learning_rate=0.005)


plt.figure(figsize=(20, 10))

plt.title('Effect of learning rate', fontsize=20)
plt.xlabel('No. of iterations', fontsize=20)
plt.ylabel('Cost', fontsize=20)
plt.xlim(1, n)

no_of_iter = list(range(0,101))

# Plotting low learning rate :
plt.plot(no_of_iter, f(np.array(low_learning_rate[1])),
         color='darkturquoise', linewidth=5, label='Low Learning Rate')

plt.scatter(no_of_iter, f(np.array(low_learning_rate[1])), s=100, color='darkturquoise')


# Plotting mid learning rate :
plt.plot(no_of_iter, f(np.array(mid_learning_rate[1])),
         color='hotpink', linewidth=5,  label='Mid Learning Rate')

plt.scatter(no_of_iter, f(np.array(mid_learning_rate[1])), s=100, color='hotpink')


# Plotting high learning rate :
plt.plot(no_of_iter, f(np.array(high_learning_rate[1])),
         color='orange', linewidth=5,  label='High Learning Rate')

plt.scatter(no_of_iter, f(np.array(high_learning_rate[1])), s=100, color='orange')

plt.legend()
```
